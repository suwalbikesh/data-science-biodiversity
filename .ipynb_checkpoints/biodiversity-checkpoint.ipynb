# Getting Started with the Data
We will begin by importing the necessary libraries and reading in the datasets:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
species = pd.read_csv('species_info.csv')
observations = pd.read_csv('observations.csv')

# Preview the data
species.head()
```

```python
observations.head()
```

Let’s explore some basic statistics and structure of the dataset.

```python
# Summary of species data
species.info()
```

```python
species['category'].value_counts()
```

```python
species['conservation_status'].value_counts(dropna=False)
```

- Many species have no listed conservation status (i.e., not considered at risk).

- The most common species types include birds, mammals, and reptiles.

## Visual and Statistical Analysis
### Distribution of Conservation Status for Animals

```python
# Fill NaN with 'No Intervention'
species['conservation_status'].fillna('No Intervention', inplace=True)
```

```python
# Plot conservation status distribution
plt.figure(figsize=(8,6))
sns.countplot(data=species, x='conservation_status', order=species['conservation_status'].value_counts().index)
plt.title('Distribution of Species by Conservation Status')
plt.ylabel('Number of Species')
plt.xlabel('Conservation Status')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

- Most species are not currently under intervention.

- A considerable number are Species of Concern, followed by a small number of Threatened or Endangered species.

## Are Certain Types of Species More Likely to Be Endangered?

```python
# Group by category and conservation status
status_counts = species.groupby(['category', 'conservation_status'])['scientific_name'].count().unstack().fillna(0)
```

```python
# Normalize to get proportions
status_props = status_counts.div(status_counts.sum(axis=1), axis=0)
```

```python
# Plot as a stacked bar chart
status_props.plot(kind='bar', stacked=True, figsize=(10,6), colormap='viridis')
plt.title('Proportion of Conservation Status by Species Category')
plt.ylabel('Proportion of Species')
plt.xlabel('Species Category')
plt.xticks(rotation=90)
plt.legend(title='Conservation Status')
plt.tight_layout()
plt.show()
```

- Mammals have a higher proportion of species under threat compared to other categories.

- Reptiles and amphibians also show notable vulnerability in specific parks.

### Statistical Significance: Are Mammals More Endangered?

```python
# Create endangered flag
species['is_protected'] = species['conservation_status'] != 'No Intervention'

# Contingency table: Mammal vs. Non-Mammal and Protected vs. Not
contingency = pd.crosstab(species['category'] == 'Mammal', species['is_protected'])

from scipy.stats import chi2_contingency
chi2, pval, dof, expected = chi2_contingency(contingency)

print("Chi-square test p-value:", pval)
```

#### Result:
If the p-value < 0.05, we reject the null hypothesis.

#### Insight:
A significant p-value indicates that mammals are statistically more likely to be endangered compared to non-mammals.

## Which Species Were Spotted the Most at Each Park?

```python
# Merge species with observations
merged = pd.merge(observations, species, on='scientific_name')

# Group and sort to find top species per park
top_species_park = merged.groupby(['park_name', 'scientific_name'])['observations'].sum().reset_index()
top_per_park = top_species_park.sort_values('observations', ascending=False).groupby('park_name').head(1)

# Plot
plt.figure(figsize=(10,6))
sns.barplot(data=top_per_park, y='park_name', x='observations', hue='scientific_name', dodge=False, palette='Set2')
plt.title('Most Spotted Species in Each Park')
plt.xlabel('Number of Observations')
plt.ylabel('Park')
plt.legend(title='Species', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

- Some species are very commonly observed, but that does not necessarily reflect their conservation status.

- Conservation priority may not align with visibility.
